```python
import random
from matplotlib import pyplot as plt

ITERATIONS = 1000

best_asst = -0.1
asst_hired_arr = []
interv_count = 10

for k in range(ITERATIONS):
    asst_hired_arr.append(0)
    best_asst = -0.1
    for i in range(interv_count):
        cur_asst = random.uniform(0, 1)
        if cur_asst > best_asst:
            asst_hired_arr[k] += 1
            best_asst = cur_asst
            
    interv_count += 1
        
plt.plot(range(10, interv_count), asst_hired_arr)
plt.yscale('log')
plt.show()
```

```python
import numpy as np

ITERATIONS = 1000

num_of_hats = 1000
hats_collection = numpy.array(list(range(num_of_hats)))

correct_arr = []
for i in range(ITERATIONS):
    new_perm = np.random.permutation(hats_collection)
    correct_count = 0
    for j in range(len(new_perm)):
        if new_perm[j] == j:
            correct_count += 1
    correct_arr.append(correct_count)
    
print(np.sum(correct_arr) / len(correct_arr))
```
